```python
import pandas as pd 
data =pd.read_csv("/content/drive/MyDrive/Colab Notebooks/RAWDATA/creditcard.csv") #loading data
data
data.head(5) #display top 5 entries
data.tail(10) #display bottom 10 entries
data.shape #displays the total number of records,rows and columns
data.describe() #give an overview summary of the dataset
data.isnull() # displays all the null entries in rows/column

```

```python
not_fraud = len(data[data.Class ==0]) #find all entries that are not fraud
not_fraud
fraud = len(data[data.Class ==1]) #find all entries that are fraud related
fraud 
not_fraud_pct = (not_fraud/(not_fraud+fraud))*100 #shows the percentage of data not fraudulent
not_fraud_pct  
fraud_pct = (fraud/(not_fraud+fraud))*100 #shows the percentage of data that is fraudulent
fraud_pct  
total_pct = not_fraud_pct+fraud_pct #shows the total percentage
total_pct

                   
```

```python
from sklearn.preprocessing import StandardScaler #normalizes the data
scaler = StandardScaler()
```

```python
data['Normalized_amount']=scaler.fit_transform(data["Amount"].values.reshape(-1,1)) #normalizes the data
data.head(4)
data.drop(["Amount","Time"],inplace=True,axis=1)
data.head(4)
```

```python
x = data.drop(["Class"],axis=1) # independent variables
y=data["Class"] #for dependent variables
x
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2) #testing and training x and y variables
```

```python
x_train.shape # number of records we have in training
x_test.shape # number of records we have in test

y_train.shape # number of records we have in training
y_test.shape # number of records we have in test
```

Creating the model:will create the model on train then test.

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
rf1 = RandomForestClassifier(n_estimators= 50) # The 50 is the total number of trees to be created.
```

```python
rf1.fit(x_train,y_train) #passing and training the datasets
```

```python
prediction_rf = rf1.predict(x_test) #predicting the model test dataset (x_test)
rf1_score = rf1.score(x_test,y_test)*100
rf1_score
```
